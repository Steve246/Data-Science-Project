The marketing team has sourced you with historical sales volumes per capita for several different drinks types.

- "year" - year (1998-2016)
- "region" - name of a federal subject of Russia. It could be oblast, republic, krai, autonomous okrug, federal city and a single autonomous oblast
- "wine" - sale of wine in litres by year per capita
- "beer" - sale of beer in litres by year per capita
- "vodka" - sale of vodka in litres by year per capita
- "champagne" - sale of champagne in litres by year per capita
- "brandy" - sale of brandy in litres by year per capita

```python
# investigate invididual regions with missing data

# visualize missing data as white strips for 'region' = 'chechen republic'
df_chechen = df[df['region'] == 'Chechen Republic']
df_chechen_yearsort = df_chechen.sort_values('year')
msno.matrix(df_chechen_yearsort)
plt.title('Missing data where region = "Chechen Republic"')

# visualize missing data as white strips for 'region' = 'Sevastopol'
df_sev = df[df['region'] == 'Sevastopol']
msno.matrix(df_sev.sort_values('year'))
plt.title('Missing data where region = "Sevastopol"')

# visualize missing data as white strips for 'region' = 'Republic of Crimea'
df_cri = df[df['region'] == 'Republic of Crimea']
msno.matrix(df_cri.sort_values('year'))
plt.title('Missing data where region = "Republic of Crimea"')

# visualize missing data as white strips for 'region' = 'Republic of Ingushetia'
df_ing = df[df['region'] == 'Republic of Ingushetia']
msno.matrix(df_ing.sort_values('year'))
plt.title('Missing data where region = "Republic of Ingushetia"')
```

Chencen Republic has an empty so we will drop from data

```python
mask = df['region'] == 'Chechen Republic'
df_no_chechen = df[~mask]
```

Drop sevasotopol and republic of crimea data but preserve the alcohol data before 2014

```python
region1 = (df_no_chechen['region'] == 'Sevastopol') | (df_no_chechen['region'] == 'Republic of Crimea') 
year1 = df_no_chechen['year'] < 2014
df_no_sev_cri = df_no_chechen[~(region1&year1)]
```

```python
# keep values from Republic of Ingushetia with complete rows and rows with only brandy data missing 
mask1 = (df_no_sev_cri['region'] == 'Republic of Ingushetia')
mask2 = (df_no_sev_cri['year']<2001) | (df_no_sev_cri['year']>2007)
df_no_ing = df_no_sev_cri[~(mask1 & mask2)]
```

```python
# input brandy data
mask1 = (df_no_ing['region'] == 'Republic of Ingushetia')
mask2 = (df_no_ing['brandy'].isnull())
df_no_ing.loc[(mask1 & mask2), 'brandy'] = 0.8
df_no_na = df_no_ing
```

```python
df_no_na.isna().sum()
```

Outlier of the data

```python
# Create tidy data for boxplot
id_vars = ['year', 'region']
value_vars = ['wine', 'beer', 'vodka', 'champagne', 'brandy']
df_no_na_tidy = df_no_na.melt(id_vars=id_vars, value_vars=value_vars, var_name='type')

# Create Boxplot
plt.figure(figsize=(14,6))
sns.boxplot(x='value', y='type', data=df_no_na_tidy)
plt.title('Distribution of sales by alcohol type')
plt.xlabel('Sales in litres by year per capita')
plt.ylabel('Alcohol type')
```

```python
# Create a separate Boxplot just for those two column
plt.figure(figsize=(14,6))
sns.boxplot(x='value', y='type', data=df_no_na_tidy[df_no_na_tidy['type'].isin(['wine', 'champagne', 'brandy'])])
plt.title('Distribution of sales by alcohol type')
plt.xlabel('Sales in litres by year per capita')
plt.ylabel('Alcohol type')
```
